```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
movies=pd.read_csv('/content/movies.csv')
ratings=pd.read_csv('/content/ratings.csv')
```

```python
movies.head()
```

```python
ratings.head()
```

```python
data=pd.merge(ratings, movies, on='movieId')
data.head()
```

```python
movie_user_matrix=data.pivot_table(index='movieId',columns='userId',values='rating')
```

```python
movie_user_matrix.fillna(0,axis=1,inplace=True)
```

```python
movie_user_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim=cosine_similarity(movie_user_matrix)
cos_sim.shape
```

```python
movie_index=pd.Series(range(len(movie_user_matrix)),index=movie_user_matrix.index)
```

```python
def recommend_movies_item_based(movie_id, n=5):
 idx=movie_index[movie_id]
 sim_scores=list(enumerate(cos_sim[idx]))
 sim_scores=sorted(sim_scores,key=lambda x:x[1],reverse=True)
 sim_scores=sim_scores[1:n+1]
 similar_movie_ids=[movie_user_matrix.index[i[0]] for i in sim_scores]
 return movies[movies['movieId'].isin(similar_movie_ids)][['movieId', 'title']]
```

```python
recommend_movies_item_based(9)
```
